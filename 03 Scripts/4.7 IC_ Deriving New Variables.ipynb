## Using If-Statements 

### With User-Defined Functions

```python
# Define the price label function

def price_label(row): 
    
    if row['prices'] <= 5:                                   
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):     
        return 'Mid-range product'
    elif row['prices'] > 15:                                   
        return 'High-range product'
    else : return 'Not enough data'
    

# If the item’s price is lower than or equal to $5, it will be labeled a “low-range product”
# If the item’s price is above $5 but lower than or equal to $15, it will be labeled a “mid-range product”
# If the item’s price is above $15, it will be labeled a “high-range product”
```

```python
# Apply the price label functionon the dataframe

df['price range'] = df.apply(price_label, axis = 1)
```

```python
# Check value counts

df['price range'].value_counts(dropna = False)
```

```python
# The most expensive product within the subset

df['prices'].max()
```

### With loc() Function

```python
# If the item’s price is above $15, it will be labeled a “high-range product”

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

### With For-Loops

```python
# Check value counts 

# 0 - Saturday , 1 - Sunday, 2 - Monday, 3 - Tuesday, 4 - Wednesday, 5 - Thursday, 6 - Friday 

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)

# Frequency of value '0' is the highest, which means Saturday is the busiest day of the week 
# Frequency of value '4' is the lowest, which means Wednesday is the slowest (least busy) day of the week
```

**Create a new column called 'Busiest day', which will contain three different values : 'Busiest day', 'Least busy', and 'Regularly busy'**


```python
# Create an empty list called result

# Set loop

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# Combine the list with the newly created column

df_ords_prods_merged['busiest_day'] = result
```

```python
# Check value counts

df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Task 2 : New column for busiest days

# Task 4 : Busiest period of day

```python
df_ords_prods_merged['order_hour_of_day'].value_counts()
```

Most orders = 10, 11, 14, 15, 13, 12, 16, 9, 17 (orders above 2000000) </br>
Average orders = 8, 18, 19, 20, 7, 21, 22 (orders above 500000) </br>
Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3 

```python
# Create an empty list called hod (hour of day)

# Set loop

hod = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9, 17] :
        hod.append("Most orders")
    elif value in [ 8, 18, 19, 20, 7, 21, 22]:
        hod.append("Average orders")
    else:
        hod.append("Fewest orders")
```

```python
# Combine the list with the newly created column

df_ords_prods_merged['busiest_period_of_day'] = hod
```

```python
# Verify if the results are correct

df_ords_prods_merged[['order_hour_of_day', 'busiest_period_of_day']].tail()
```
