## Cosistency Checks

Exercise 4.9 - Step 5

```python
# Checking data types

df_customers.dtypes
```

```python
# user_id is an unique identification number, which can be converted to str

df_customers['user_id'] = df_customers['user_id'].astype('str')
```

```python
# Checking the data type of user_id

df_customers['user_id'].dtypes
```

```python
# Checking the descriptive statistics 

df_customers.describe()
```

```python
# Checking for missing values

df_customers.isnull().sum()
```

```python
# Check for mixed-types

# Function for checking whether a datframe contains any mixed-type columns :

for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)   # Weird (boolean value) : a new variable is created, which checks if the data types within the column are consistent
    if len (df_customers[weird]) > 0:          # 0 : False and 1 : True; So if the value is True , i.e 1 ,then it will print the column name
        print(col)
        
```

No column names were printed, which implies no columns in df_customers are mixed-type columns.

```python
# Checking for duplicates

df_customers_dups = df_customers[df_customers.duplicated()]
```
