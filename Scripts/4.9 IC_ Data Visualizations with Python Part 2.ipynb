#### Sampling

```python
# Create a seed
# Set the seed for the random number generator
# Create 32,404,859 random numbers (the length of your df_combined dataframe) ranging between 0 and 1
# Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7)
# Store the results of the test in a new list, dev, in the form of boolean values (True and False)

np.random.seed(4)
dev = np.random.rand(len(df_combined)) < 0.7
```

```python
dev
```

```python
np.random.rand(10)
```

```python
# Split the dataframe into two samples

big = df_combined[dev]
small = df_combined[~dev]

# big sample contains 70% of the dataframe values 
# small sample contains the reamaining 30% of the values
```

```python
# Length of df_combined

len(df_combined)
```

```python
# Verifying the length

len(big) + len (small)
```

```python
# Reducing the samples to necessary columns, to help speed up the proccessing

df_2 = small[['orders_day_of_week', 'prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

The line is at a peak at 0 and 6, representing that most money is spent on Saturday and Sunday.

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_dow_prices.png'))
```
